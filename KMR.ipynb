```julia
include("KMR2.jl")
print()
```

```julia
M = Array(Int, (2, 2, 2))
M[1, 1, :] = [4, 4]
M[1, 2, :] = [0, 3]
M[2, 1, :] = [3, 0]
M[2, 2, :] = [2, 2]

N = 10
T = 10^7
epsilon = 0.01
```

```julia
kmr = KMR(M, N, epsilon)
```

```julia
@time simulate!(kmr, T)
print()
```

```julia
plot_sample_path(kmr)
```

```julia
plot_empirical_dist(kmr)
```

```julia
plot_stationary_dist(kmr)
```
